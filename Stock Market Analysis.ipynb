<b>1. What are the top 5 best and worst-performing stocks in the last year based on percentage change?</b>

```python
df=df[['Company Name','Symbol','Previous Close', 'Last Traded Price', '365 Day Percentage Change']]
df
```

```python
# Convert '365 Day Percentage Change' to numeric type
df['365 Day Percentage Change'] = pd.to_numeric(df['365 Day Percentage Change'], errors='coerce')

# Drop any rows with NaN values after conversion
df.dropna(subset=['365 Day Percentage Change'], inplace=True)

# Sort stocks by percentage change
top_5_best = df.nlargest(5, '365 Day Percentage Change')
top_5_worst = df.nsmallest(5, '365 Day Percentage Change')

# Display results
print("Top 5 Best Performing Stocks:")
print(top_5_best[['Company Name', 'Symbol', '365 Day Percentage Change']])

print("\nTop 5 Worst Performing Stocks:")
print(top_5_worst[['Company Name', 'Symbol', '365 Day Percentage Change']])
```

<b>	2. How does the daily stock price movement (Open, High, Low, Close) fluctuate over time?</b>

```python
df=pd.read_csv('nifty_500.csv')
df
```

```python
DailyStockPrice=df[['Company Name','Symbol','Open','High','Low','Last Traded Price']].head()
DailyStockPrice
```

```python
import matplotlib.pyplot as plt

# Extract relevant columns
x = DailyStockPrice['Company Name']
y_open = DailyStockPrice['Open']
y_high = DailyStockPrice['High']
y_low = DailyStockPrice['Low']
y_last_traded = DailyStockPrice['Last Traded Price']

# Set figure size
fig, axs = plt.subplots(2, 2, figsize=(15, 10))  # Create a 2x2 grid of subplots

# Plot Open Price
axs[0, 0].plot(x, y_open, linestyle='dotted', color='blue', marker='o')
axs[0, 0].set_title('Open Price')

# Plot High Price
axs[0, 1].plot(x, y_high, linestyle='dotted', color='yellow', marker='s')
axs[0, 1].set_title('High Price')

# Plot Low Price
axs[1, 0].plot(x, y_low, linestyle='dotted', color='green', marker='^')
axs[1, 0].set_title('Low Price')

# Plot Last Traded Price
axs[1, 1].plot(x, y_last_traded, linestyle='dotted', color='red', marker='*')
axs[1, 1].set_title('Last Traded Price')

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

<b>	3. Which stocks have shown the highest volatility (using standard deviation of returns)?</b>

```python
import pandas as pd

# Ensure 'Last Traded Price' and 'Previous Close' are numeric
df['Last Traded Price'] = pd.to_numeric(df['Last Traded Price'], errors='coerce')
df['Previous Close'] = pd.to_numeric(df['Previous Close'], errors='coerce')

df['Daily Returns'] = (df['Last Traded Price'] - df['Previous Close']) / df['Previous Close']

volatility = df.groupby('Company Name')['Daily Returns'].std().reset_index()

volatility_sorted = volatility.sort_values(by='Daily Returns', ascending=False)

print(volatility_sorted.head(10))
```

<b>4. What is the distribution of stock returns (Histogram/KDE Plot)?</b>

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate stock returns
DailyStockPrice['Daily Returns'] = (df['Last Traded Price'] - df['Previous Close']) / df['Previous Close']

# Plot Histogram and KDE
plt.figure(figsize=(10,5))
sns.histplot(df['Daily Returns'], kde=True, bins=50, color='blue')
plt.title("Distribution of Stock Returns")
plt.xlabel("Daily Returns")
plt.ylabel("Frequency")
plt.show()
```

<b>5. How do stock returns correlate with each other (Correlation Heatmap)?</b>

```python
import numpy as np

# Compute correlation matrix
corr_matrix = df[['Open', 'High', 'Low', 'Previous Close', 'Last Traded Price']].corr()

# Plot heatmap
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation between Stock Prices")
plt.show()
```

<b>7. How do moving averages (SMA, EMA) compare to actual stock prices?</b>

```python
df['SMA_50'] = df['Last Traded Price'].rolling(window=50).mean()
df['EMA_50'] = df['Last Traded Price'].ewm(span=50, adjust=False).mean()

plt.figure(figsize=(12,6))
# plt.plot(df['Date'], df['Last Traded Price'], label="Stock Price")
plt.plot(df['Previous Close'], df['SMA_50'], label="50-day SMA", linestyle='dashed')
plt.plot(df['Previous Close'], df['EMA_50'], label="50-day EMA", linestyle='dotted')
plt.legend()
plt.title("Stock Price vs SMA vs EMA")
plt.show()
```

<b>8. Which sectors have seen the highest growth and decline in the last 30/90/365 days?</b>

```python
df['365d Change'] = ((df['Last Traded Price'] - df['Last Traded Price'].shift(365)) / df['Last Traded Price'].shift(365)) * 100

sector_growth = df.groupby('Industry')['365d Change'].mean().sort_values(ascending=False)
print(sector_growth.head(10))  # Top 10 growing sectors
print(sector_growth.tail(10))  # Top 10 declining sectors
```

<b>10. Compute the Sharpe Ratio to evaluate risk-adjusted returns</b>

```python
risk_free_rate = 0.02
df['Daily Returns'] = df['Last Traded Price'].pct_change()
sharpe_ratio = (df['Daily Returns'].mean() - risk_free_rate / 252) / df['Daily Returns'].std()
print(f"Sharpe Ratio: {sharpe_ratio:.2f}")
```

<b>11. Analyze the Max Drawdown (largest peak-to-trough decline)</b>

```python
df['Cumulative Max'] = df['Last Traded Price'].cummax()
df['Drawdown'] = (df['Last Traded Price'] - df['Cumulative Max']) / df['Cumulative Max']
max_drawdown = df['Drawdown'].min()
print(f"Max Drawdown: {max_drawdown:.2%}")
```
